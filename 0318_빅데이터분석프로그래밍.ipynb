### 내장함수 arange()와 linspace()

```python
import numpy as np

print(np.arange(3)) # 0~2까지
print(np.arange(3, 7))
print(np.arange(3, 7, 2)) # 스템, 2씩 증가
```

```python
print(np.arange(2.0))
print(np.arange(2.0, 6.5))
print(np.arange(2.0, 6.5, 0.8))
```

```python
a = np.arange(12).reshape(3, 4)
a
```

```python
np.reshape(a, (2, 6))
```

```python
# np.reshape(a, (3, 5)) 원소가 15개인데 원래 배열 원소는 12개이기 때문에 오류
```

```python
np.linspace(2.0, 3.0)
```

```python
np.linspace(2.0, 3.0, retstep=True)
```

```python
print( (3-2) / (50-1))
```

```python
np.linspace(2, 3, 3, retstep = True)
```

```python
a = np.linspace(2, 3, 3, retstep = True)
float(a[1])
```

```python
np.linspace(2, 3, 3, retstep = True, endpoint = False)
```

```python
(3-1) / 3
```

### sum

```python
a = np.array(([3, 5, 8], [1, 2, 3]))
```

```python
# 일차원으로 펼쳐서 계산
a.cumsum()
```

```python
a.cumsum(axis=0) # 시험예상, 행을 따라 세로로 합산
```

```python
a.cumsum(axis=1) # 시험예상, 행을 따라 가로로 합산
```
